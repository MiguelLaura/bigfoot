```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
X = pd.read_csv('data/bigfoot_2.csv')
X.info()
```

```python
X_nb = X[
    [
        "latitude",
        "longitude",
        "humidity",
        "cloud_cover",
        "moon_phase",
        "precip_probability",
        "pressure",
        "uv_index",
        "wind_bearing",
        "temperature_high_deg",
        "temperature_mid_deg",
        "temperature_low_deg",
        "dew_point_deg",
        "wind_speed_kmh",
        'precip_intensity_cm',
        'visibility_km'
    ]
]
X_nb.dropna(inplace=True)
X_nb.info()
```

```python
cls = PCA()
pcs = cls.fit_transform(X_nb)
plt.bar(range(cls.components_.shape[0]), cls.explained_variance_ratio_)
plt.show()
```

```python
cls.components_[0]
```

```python
loadings = pd.DataFrame(cls.components_, columns=X_nb.columns)
loadings
```

```python
def scale(data):
    scaler = StandardScaler()
    scaler.fit(data)
    tab = scaler.transform(data)
    return tab
```

```python
X_norm = scale(X_nb)
```

```python
cls = PCA()
pcs = cls.fit_transform(X_norm)
plt.bar(range(cls.components_.shape[0]), cls.explained_variance_ratio_)
plt.show()
```

```python
cls.components_[0]
```

```python
cls.components_[1]
```

```python
plt.figure(figsize=(20,8))
df_X = pd.DataFrame(pcs, columns=[f"PC{i}" for i in range(cls.components_.shape[0])])
sns.scatterplot(x="PC1", y="PC2", hue=X.state, size=X.season, data=df_X)
```
